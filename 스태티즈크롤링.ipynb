```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
#크롬 드라이버 버전 확인 Chrome://version

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chromeDriver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
chromeDriver.get("http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=0&ys=2022&ye=2022&se=0&te=%EB%91%90%EC%82%B0&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=0&tr=&cv=&ml=1&sn=30&si=&cn=")
#사용할 URL

html = chromeDriver.page_source
bsObject = BeautifulSoup(html, 'html.parser')
```

```python
temp = bsObject.find_all("table")[1]
temp
```

```python
column = ["순","이름","팀","WAR*","G","타석","타수","득점","안타","2타","3타","홈런","루타","타점",
          "도루","도실","볼넷","사구","고4","삼진","병살","희타","희비","타율",
          "출루","장타","OPS","wOBA",'wRC+','WAR','WPA']
df = pd.DataFrame(columns=column)
templen = len(temp.find_all("tr"))

for i in range(2, templen):
  tempTr = temp.find_all("tr")[i]
  if(tempTr.find("th") is not None):
    continue
  row = {}
  column_idx = 0
  for j in range(31):
    tempTd = tempTr.find_all("td")[j].text
    row[column[column_idx]] = tempTd
    column_idx += 1
  df = df.append(row,ignore_index=True)
```

```python
df = df.iloc[:22]
df
```

```python
df.to_csv('2022_두산_타자_기본.csv')
```
