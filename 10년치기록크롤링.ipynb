```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get("http://www.statiz.co.kr/league.php?opt=2022&sopt=1")

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

tables = soup.find_all('table')

for i, table in enumerate(tables):
    data = []
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all('td')
        data.append([cell.text for cell in cells])

    df = pd.DataFrame(data)
    print(f"DataFrame for table {i}:")
    df.to_csv(f'22두산_{i}.csv')

driver.quit()
```

```python
team_basic = pd.DataFrame(columns = ['순', '이름', '팀', 'WAR*', 'G', '타석', '타수', '득점', '안타', '2타', '3타', '홈런',
       '루타', '타점', '도루', '도실', '볼넷', '사구', '고4', '삼진', '병살', '희타', '희비', '타율',
       '출루', '장타', 'OPS', 'wOBA', 'wRC+', 'WAR*', 'WPA'])
for year in range(2012,2023):
    url = f'http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=0&ys={year}&ye={year}&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&de=1&lr=5&tr=&cv=&ml=1&sn=30&si=&cn='
    year_s = pd.read_html(url)[0].iloc[4:,:31]
    year_s.columns =year_s.columns.droplevel([0,1])
    team_basic = pd.concat([team_basic,year_s])
```

```python
team_basic.reset_index(drop=True,inplace = True)
team_basic
```
