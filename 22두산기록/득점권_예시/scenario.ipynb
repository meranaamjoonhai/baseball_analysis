```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
pd.set_option('display.max_columns',None)
```

```python
df1 = pd.read_csv('1S3B_1.csv',encoding='UTF-8')
df1.drop(columns=['Unnamed: 0','순','비율.1'],inplace = True)
df1.columns = df1.iloc[0]
df1 = df1.iloc[1:,:]
df1
```

```python
df2 = pd.read_csv('1S3B_2.csv',encoding='UTF-8')

# 첫번째 행을 컬럼으로 지정
df2.columns = [col if col in df2.iloc[0].values else f"{col} ({df2.iloc[0][col]})" for col in df2.columns]

# 첫번째 행 삭제
df2 = df2.iloc[1:, :]

# 결과 확인
df2.drop(columns=['Unnamed: 0 (nan)','순'],inplace = True)
df2
```

```python
df_inner_join = pd.merge(df1,df2,on = ['이름','팀'])
df_inner_join
```

```python
df3 = pd.read_csv('2S3B_1.csv',encoding='UTF-8')
df3.drop(columns=['Unnamed: 0','순','비율.1'],inplace = True)
df3.columns = df3.iloc[0]
df3 = df3.iloc[1:,:]
df3
```

```python
df4 = pd.read_csv('2S3B_2.csv',encoding='UTF-8')

# 첫번째 행을 컬럼으로 지정
df4.columns = [col if col in df4.iloc[0].values else f"{col} ({df4.iloc[0][col]})" for col in df4.columns]

# 첫번째 행 삭제
df4 = df4.iloc[1:, :]

# 결과 확인
df4.drop(columns=['Unnamed: 0 (nan)','순'],inplace = True)
df4
```

```python
df_inner_join2 = pd.merge(df3,df4,on = ['이름','팀'])
df_inner_join2
```

```python
compare_wpa1 = df_inner_join[['이름','WPA']]
compare_wpa2 = df_inner_join2[['이름','WPA']]
```

```python
compare_wpa = pd.merge(compare_wpa1,compare_wpa2,on = ['이름'])
```

```python
compare_wpa.columns = ['이름','1S3B_WPA','2S3B_WPA']
compare_wpa[['1S3B_WPA','2S3B_WPA']] = compare_wpa[['1S3B_WPA','2S3B_WPA']].apply(pd.to_numeric)
compare_wpa
```

```python
import plotly
import plotly.graph_objs as go
import plotly.express as px 

fig = px.line(compare_wpa, x='이름', y=['1S3B_WPA','2S3B_WPA'])
fig.show()

```
